## Diagram

```python
df.drop_duplicates
Heighest=['H1','H2','H3','H4','H5']
Amount=df['Amount'].nlargest(5)
plt.plot(Heighest,Amount,color='b',marker='o',linestyle=':')
plt.xlabel('Top 5 Heighest Amount')
plt.ylabel('Actual Amount')
plt.title('User Data',fontsize='15')
plt.show()
```

```python
from matplotlib import style
df.drop_duplicates
Heighest=['H1','H2','H3','H4','H5']
Amount=df['Amount'].nlargest(5)
style.use('ggplot')
plt.plot(Heighest,Amount,color='b',marker='o',linestyle=':')
plt.xlabel('Top 5 Heighest Amount')
plt.ylabel('Actual Amount')
plt.title('User Data',fontsize='15')
plt.grid(color='y',linestyle="-",linewidth=1)
plt.show()
```

```python
Heighest=['H1','H2','H3','H4','H5']
Amount=df['Amount'].nlargest(10)
plt.hist(Amount)
plt.show
```

```python
bins=[8000,8500,9000,9500,10000,15000,20000,25000,30000]
Amount=df['Amount'].nlargest(30)
plt.hist(Amount,bins,rwidth=0.5,histtype='bar',orientation='vertical',color='c',label='Amount')
plt.title('Highest Amount Data')
plt.xlabel('User')
plt.ylabel('Amount')
plt.legend()
plt.show
```
